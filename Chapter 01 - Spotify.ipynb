### 6.5. Removing outliers outside the range

#### 6.5.1. Function to calculate quartiles of column and remove outliers outside range

```python
def drop_outliers_IQR(column):
    
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)

    IQR = Q3 - Q1
    upper_limit = Q3 + 1.5 * IQR
    lower_limit = Q1 - 1.5 * IQR

    outliers = column[(column > upper_limit) | (column < lower_limit)]
    
    outliers_dropped_column = column.drop(outliers.index)
    
    return outliers_dropped_column
```

#### 6.5.2. Calling the function with columns and reassigning with latest data

```python
# key, acousticness, valence does not have any outliers
outlier_columns = ['Danceability', 'Energy', 'Loudness', 'Speechiness', 'Liveness', 'Tempo']

for i in range(0, 11):
    for columns in outlier_columns:
        result_column = drop_outliers_IQR(df[columns])
        df[columns] = result_column
```
